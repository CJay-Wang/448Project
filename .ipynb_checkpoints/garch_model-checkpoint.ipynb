```python
import pandas as pd
from arch import arch_model
# from arch import ConstantMean, GARCH, Normal
import numpy as np
import random
```

```python
price = pd.read_csv('aapl20150102price1hr.txt', header = None, names = ['Datetime','Price'], index_col = 0)
price.head()
```

```python
# plot prices

%matplotlib inline
import seaborn
seaborn.mpl.rcParams['figure.figsize'] = (10.0, 6.0)
seaborn.mpl.rcParams['savefig.dpi'] = 90
price.plot()
```

```python
# plot log_returns

# log_return = log(pi/pj)
log_pj = list(np.log(price['Price']))
log_pi = log_pj[1:]
log_pi.append(0)
assert len(log_pi) == len(log_pj)
price['log_ret'] = np.array(log_pi) - np.array(log_pj)
price = price[:-1]
price.plot(y = 'log_ret')
```

```python
# GARCH model
returns = 10000 * np.array(price['log_ret']) # need to scale to prevent vanishing gradient during optimization
print 'after scaling returns by 10000: ', returns[:20]
am = arch_model(returns)

# am = ConstantMean(returns)
# am.volatility = GARCH(1, 0, 1)
# am.distribution = Normal()
res = am.fit()
```

```python
print(res.summary())
```

```python
fig = res.plot(annualize='D')
```
